```python

import pandas as pd
url = 'https://barttorvik.com/trank.php?year=2023&sort=&top=0&conlimit=All&venue=All&type=All#'
df = pd.read_html(url,header=1) 
```

```python
CYData = df[0]
CYData = CYData.head(-1) #Removing last row because it is a summary of stats above
CYData
```

```python
CYData[['Rec', 'Conf Record']] = CYData['Rec'].str.split(' ', expand=True)
CYData[['AdjOE', 'AdjOE Rank']] = CYData['AdjOE'].str.split(' ', expand=True)
CYData[['AdjDE', 'AdjDE Rank']] = CYData['AdjDE'].str.split(' ', expand=True)
CYData[['Barthag', 'Barthag Rank']] = CYData['Barthag'].str.split(' ', expand=True)
CYData[['EFG%', 'EFG% Rank']] = CYData['EFG%'].str.split(' ', expand=True)
CYData[['EFGD%', 'EFGD% Rank']] = CYData['EFGD%'].str.split(' ', expand=True)
CYData[['TOR', 'TOR Rank']] = CYData['TOR'].str.split(' ', expand=True)
CYData[['ORB', 'ORB Rank']] = CYData['ORB'].str.split(' ', expand=True)
CYData[['DRB', 'DRB Rank']] = CYData['DRB'].str.split(' ', expand=True)
CYData[['FTR', 'FTR Rank']] = CYData['FTR'].str.split(' ', expand=True)
CYData[['FTRD', 'FTRD Rank']] = CYData['FTRD'].str.split(' ', expand=True)
CYData[['2P%', '2P% Rank']] = CYData['2P%'].str.split(' ', expand=True)
CYData[['2P%D', '2P%D Rank']] = CYData['2P%D'].str.split(' ', expand=True)
CYData[['3P%', '3P% Rank']] = CYData['3P%'].str.split(' ', expand=True)
CYData[['3P%D', '3P%D Rank']] = CYData['3P%D'].str.split(' ', expand=True)
CYData['Team'] = CYData['Team'].str.partition('(')[0]
CYData
```

```python
#pip install matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
CYData = CYData[CYData['FTR Rank'].notnull()] #Getting rid of the rows spacer rows
CYData['Rk'] = pd.to_numeric(CYData['Rk'])
CYData['EFG%'] = pd.to_numeric(CYData['EFG%'])
CYData['EFGD%'] = pd.to_numeric(CYData['EFGD%'])
CYData['AdjOE'] = pd.to_numeric(CYData['AdjOE'])
CYData['AdjDE'] = pd.to_numeric(CYData['AdjDE'])
```

```python

Top100Teams = CYData.query('Rk < 25')
ax = Top100Teams.plot.scatter(x='AdjOE', y = 'AdjDE')
for idx, row in Top100Teams.iterrows():
    ax.annotate(row['Team'], (row['AdjOE'],row['AdjDE']))
ax.axvline(np.percentile(Top100Teams['AdjOE'],50), color='red')
ax.axhline(np.percentile(Top100Teams['AdjDE'],75), color='red')
```
