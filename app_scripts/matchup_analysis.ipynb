```python
import pandas as pd
```

```python
df_monte = pd.read_csv('matchups_sim_results.csv')
```

```python
df_monte.head()
```

```python
df_monte = df_monte[['home_team','away_team','mean_spread_away_minus_home','home_win_prob']]

df_monte.dropna(inplace=True)

df_monte
```

```python
df_bart = pd.read_csv('matchups_bart_results.csv')
df_bart.head()
```

```python
df_bart = df_bart[['home_team','away_team','mean_spread_away_minus_home']]
df_bart.dropna(inplace=True)
df_bart.rename(columns={'mean_spread_away_minus_home':'spread_away_minus_home_bart'}, inplace=True)
df_bart
```

```python
df_final = pd.merge(df_monte, df_bart, on=['home_team','away_team'])
df_final['mean_spread_away_minus_home'] = df_final['mean_spread_away_minus_home']-3
df_final
```

```python
df_final['avg_result'] = (df_final['mean_spread_away_minus_home'] + df_final['spread_away_minus_home_bart']) / 2
```

```python
odds = pd.read_csv('Data/odds_today.csv')
odds = odds.rename(columns={'Team Modified':'team'})
odds = odds[(odds['Bookmaker']=='fanduel')]
odds = odds[odds['Market']=='spreads']
odds.head()
```

```python
df_final[df_final['away_team'] == 'Boise St.']
```

```python
df_final_odds = pd.merge(df_final, odds[['team','Point']], left_on=['home_team'], right_on=['team'])
```

```python
df_final_odds.head()
```

```python
df_final_odds.to_csv('matchup_analysis_with_odds_0113_2.csv', index=False)
```
